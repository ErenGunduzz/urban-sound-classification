### Normalizing and grayscaling the data

```python
new = []

for i in data_dir:
    imgnpy = np.load(i)
    grayimg = cv2.cvtColor(imgnpy,cv2.COLOR_BGR2GRAY)
    new.append(grayimg)
newest =[]
for file in new:
    normfile = file/255
    newest.append(normfile)

    
```

```python
newest[0].shape
```

```python
array = np.array(newest)
array.shape
```

```python
plt.hist(array.ravel(), bins=50, density=True)
plt.xlabel("pixel values")
plt.ylabel("relative frequency")
plt.title("distribution of pixels")
print(array)
```

```python
train = array.reshape(array.shape[0], array.shape[1],array.shape[2],1)
print(train.shape)
print(train[0].shape)
np.save(r"C:\Users\asus\Desktop\spectrograms\X.npy",train)
```
